## Finding the position vector
If P is the point with coordinates (x, y, z) then the length, or magnitude, of the position vector
OP is given by the formula: 
### |OP|= sqrt(x^2 + y^2 + z^2)

```python
def square(x):
    return x**2

def positionVector(x, y, z):
    x = square(x)
    y = square(y)
    z = square(z)
    return(abs(sqrt(x + y+ z))) 

final_df = pd.DataFrame(columns=['sumGyro', 'sumAcc', 'sumMag'])
for index, row in initial_df.iterrows():
    final_df.at[index,'sumGyro'] = positionVector(row['GyroX'], row['GyroY'], row['GyroZ']) #get OP of Gyro
    final_df.at[index,'sumAcc'] = positionVector(row['AccX'], row['AccY'], row['AccZ']) #get OP of Acc
    final_df.at[index,'sumMag'] = positionVector(row['MagX'], row['MagY'], row['MagZ']) #get OP of Mag
```

```python
print final_df 
```

```python
print "sum of Gyro: "+ str(final_df['sumGyro'].sum())
print "sum of Acc: "+ str(final_df['sumAcc'].sum())
print "sum of Mag: "+ str(final_df['sumMag'].sum())
```
